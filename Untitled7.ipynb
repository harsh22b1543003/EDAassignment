```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Load the Titanic dataset
titanic_df = pd.read_csv('tested.csv')
```

```python
#Calculate survival rate by passenger class
survival_rate_by_class = titanic_df.groupby('Pclass')['Survived'].mean().reset_index()
```

```python
# Plot the survival rate by passenger class
plt.figure(figsize=(10, 5))
sns.barplot(x='Pclass', y='Survived', data=survival_rate_by_class)
plt.title('Survival Rate by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')
plt.show()
```
